```python
File = open(r'C:\Users\Chris.Cirelli\Desktop\Python Programming Docs\GSU\Sprint Project\Docket-Sheet-Classification\Code_Algorithm\Moby_Dick.txt')
Text = File.read()
```

```python
Text_tokenized = Text.split(' ')
```

```python
List_bigrams = []

Text_bigrams = [List_bigrams.append((x.lower(),y.lower())) for x,y in zip(Text_tokenized, Text_tokenized[1:])]
```

```python
Dict = {}

for x in List_bigrams:
    Dict[x] = Dict.get(x, 0) + 1
    
```

```python
df = pd.DataFrame(Dict, index = [0]).transpose()
```
